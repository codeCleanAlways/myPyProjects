```python
import pandas as pd
import numpy as np

dfOutcomes = pd.read_csv("sc1_Phase1_GE_Outcome.tsv",sep="\t")
dfOutcomes.describe()


```

```python
dfFeatures = pd.read_csv("sc1_Phase1_GE_FeatureMatrix.tsv",sep="\t")
dfFeatures.head()
```

```python
dfPhenotypes = pd.read_csv("sc1_Phase1_GE_Phenotype.tsv",sep="\t")
dfPhenotypes = dfPhenotypes.replace(r'^\s*$', np.nan, regex=True)
dfPhenotypes
```

```python
dfs = pd.merge(dfPhenotypes,dfOutcomes,how="inner",on="PATIENTID")
dfMerged = pd.merge(dfs,dfFeatures,how="inner",on="PATIENTID")
#dfMerged.describe()
dfMerged.head()
```

```python
from sklearn.feature_selection import GenericUnivariateSelect, chi2, SelectKBest, f_regression
X = dfMerged
X=X.drop(columns=['PATIENTID','SEX','RACE','WHO_GRADING','CANCER_TYPE','SURVIVAL_STATUS'])
y = dfMerged.iloc[:,4]
X.head()
y.head()
```

```python
bestfeatures = SelectKBest(score_func=chi2, k=100)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
#concat two dataframes for better visualization 
featureScores = pd.concat([dfcolumns,dfscores],axis=1)
featureScores.columns = ['Specs','Score']  #naming the dataframe columns
print(featureScores.nlargest(100,'Score'))  #print 10 best features
```

```python
import matplotlib.pyplot as plt
Vals= featureScores.nlargest(28,'Score')  #print 10 best features
xVal = Vals['Specs']
yVal = Vals['Score']
#plt.plot(xVal,yVal)  #print 10 best features
#plt.show()
```

```python
selCols = pd.Series(['CANCER_TYPE'])
dfSelectedY = dfMerged.loc[:,selCols]
#xVal = pd.concat([selCols,xVal])
dfSelectedX = dfMerged.loc[:,xVal]
dfSelectedX.drop(['PATIENTID','CANCER_TYPE'],axis=1,inplace=True)
dfSelectedX
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dfSelectedX, dfSelectedY, test_size=0.15, random_state=0)

from sklearn.linear_model import LogisticRegression
# all parameters not specified are set to their defaults
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train, y_train)



predictions = logisticRegr.predict(x_test)
# Use score method to get accuracy of model
#print(y_test)

predY = pd.DataFrame(predictions,index= y_test.index)
#print(predY)

finalPred = pd.concat([y_test,predY],axis=1)
print(finalPred)

score = logisticRegr.score(x_test, y_test)
print(score)
```
